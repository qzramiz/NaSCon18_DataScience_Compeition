```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline
sns.set()
```

```python
files = list()
for f in os.listdir('../TrainECGData/'):
    files.append('../TrainECGData/{}/0.csv'.format(f, f))
for f in os.listdir('../TestECGData/'):
    files.append('../TestECGData/{}/0.csv'.format(f, f))
files
```

```python
frm = list()
patid = list()
images = list()
```

```python
for fl in files:
    data = pd.read_csv(fl)
    columns = data.columns
    for total in range(data.shape[0] / 3000):
        f, ax = plt.subplots(3, 5 , sharex=True, sharey=True)
        f.set_figheight(10)
        f.set_figwidth(15)
        a = 0
        b = 0
        st = 3000 * total
        end = 3000 * (total + 1)
        for i, c in enumerate(columns):
            if i == 0:
                continue
            elif i == 1:
                ax[a,b].plot(data[c][st:end])
                b += 1
            else:
                ax[a,b].plot(data[c][st:end])
                b += 1
                if b == 5:
                    a += 1
                    b = 0
        name = str(fl.split('/')[-2]) + "_" + str(total)
        if 'train' in fl.lower():
            f.savefig('train/{}.jpg'.format(name), bbox_inches='tight')
            frm.append('train')
        else:
            f.savefig('train/{}.jpg'.format(name), bbox_inches='tight')
            frm.append('test')
        patid.append(fl.split('/')[-2])
        images.append(name+".jpg")
        plt.close(f)
    del data
    print(fl, "Done")
```
