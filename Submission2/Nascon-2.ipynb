```python
import pandas as pd
import numpy as np
pd.options.display.max_columns =100
```

```python
train = pd.read_csv('Patient_Records_Train.csv')
test = pd.read_csv('Patient_Records_Test.csv')
print(train.shape)
print(test.shape)
```

```python
train['from'] = 'train'
test['from'] = 'test'
```

```python
all_ = pd.concat([train, test.drop('Usage', axis=1)])
all_.shape
```

```python
all_.columns = all_.columns.str.lower()
all_.columns
```

```python
all_.dtypes
```

```python
all_.isnull().sum()
```

```python
dat = all_[['acute infarction (localization)', 'from']]
dat['acute infarction (localization)'] = dat['acute infarction (localization)'].str.strip()
```

```python
dat.loc[~((dat['acute infarction (localization)'] == 'n/a') |
    (dat['acute infarction (localization)'] == 'no') |
     (dat['acute infarction (localization)'] == 'unknown')), 'acute infarction (localization)'] = 'constant'
dat
```

```python
dat = pd.concat([dat, all_[['in hospital medication']]], axis=1)
dat['in hospital medication'] = dat['in hospital medication'].str.strip()
```

```python
dat.loc[~((dat['in hospital medication'] == 'n/a')), 'in hospital medication'] = 'constant'
dat
```

```python
dat = pd.concat([dat, all_[['additional diagnoses']]], axis=1)
dat['additional diagnoses'] = dat['additional diagnoses'].str.strip()
```

```python
dat.loc[dat['additional diagnoses'].str.contains('renal insufficiency'), 'additional diagnoses'] = 'infaction'
dat.loc[dat['additional diagnoses'].str.contains('aortic stenosis'), 'additional diagnoses'] = 'valve'
dat.loc[dat['additional diagnoses'].str.contains('regurgitation'), 'additional diagnoses'] = 'valve'
dat.loc[dat['additional diagnoses'].str.contains('palpitation, obesity'), 'additional diagnoses'] = 'disrythmia'
dat.loc[dat['additional diagnoses'].str.contains('palpitation, arterial hypertension'), 'additional diagnoses'] = 'infaction'
dat.loc[(dat['additional diagnoses'].str.contains('cardiomyopathy'))
        & (dat['additional diagnoses'].str.contains('ventricular')), 'additional diagnoses'] = 'cardiomyopathy'
dat.loc[dat['additional diagnoses'].str.contains('hypertrophic'), 'additional diagnoses'] = 'cardiomyopathy'
```

```python
dat.loc[dat['additional diagnoses'].str.contains('cardiomyopathy'), 'additional diagnoses'] = 'cardiomyopathy'
dat.loc[~((dat['additional diagnoses'] == 'unknown') |
    (dat['additional diagnoses'] == 'no') |
    (dat['additional diagnoses'] == 'valve') |
    (dat['additional diagnoses'] == 'disrythmia') |
    (dat['additional diagnoses'] == 'infaction') |
     (dat['additional diagnoses'] == 'cardiomyopathy')), 'additional diagnoses'] = 'constant'
dat['additional diagnoses'].value_counts()
```

```python
dat = pd.concat([dat, all_['ventriculography']], axis=1)
dat['ventriculography'] = dat['ventriculography'].str.strip()
```

```python
dat.loc[~(dat['ventriculography'].str.contains('not')), 'ventriculography'] = 'constant'
dat
```

```python
dat = pd.concat([dat, all_['former infarction (localization)']], axis=1)
dat['former infarction (localization)'] = dat['former infarction (localization)'].str.strip()
dat
```

```python
dat.loc[~((dat['former infarction (localization)'].str.contains('no')) |
          (dat['former infarction (localization)'].str.contains('unknown'))), 'former infarction (localization)'] = 'constant'
dat
```

```python
dat = pd.concat([dat, all_[['age', 'sex']]], axis=1)
dat
```

```python
dat.isnull().sum()
```

```python
dat.loc[dat['age'].isnull(), 'age'] = dat['age'].median()
dat.isnull().sum()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import train_test_split
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import f1_score
```

```python
m = RandomForestClassifier(n_jobs=4)
params = {'n_estimators': [35, 50, 75, 100, 125, 150],
          'min_samples_split': [2, 4, 8, 16, 32]}
gs = GridSearchCV(m, params)
```

```python
dat = pd.concat([all_['disease'], dat], axis=1)
```

```python
dat.dtypes
```

```python
ac = dat['acute infarction (localization)'].str.get_dummies()
ac.columns = ac.columns + '_ac'
ih = dat['in hospital medication'].str.get_dummies()
ih.columns = ih.columns + '_ih'
ad = dat['additional diagnoses'].str.get_dummies()
ad.columns = ad.columns + '_ad'
ve = dat['ventriculography'].str.get_dummies()
ve.columns = ve.columns + '_ve'
fo = dat['former infarction (localization)'].str.get_dummies()
fo.columns = fo.columns + '_fo'
dat['sex'] = dat['sex'].str.strip()
se = dat['sex'].str.get_dummies()
se.columns = se.columns + '_se'
```

```python
d = pd.concat([dat[['from', 'age', 'disease']], ac, ih, ad, ve, fo, se], axis=1)
```

```python
d.columns
```

```python
d.to_csv('engineered_rules.csv', index=False)
```

```python
d = pd.read_csv('allfeatures.csv')
```

```python
d['disease'] = d['disease'].astype('category')
d['diseasecode'] = d['disease'].cat.codes
```

```python
(np.abs(d.corr()) > .5)
```

```python
tr = d[d['from'] == 'train']
ts = d[d['from'] == 'test']
tr = tr.drop('from', axis=1)
ts = ts.drop('from', axis=1)
```

```python
X = tr.drop(['age', 'disease', 'diseasecode'], axis=1).values
Y = tr['diseasecode'].values
```

```python
xtr, xts, ytr, yts = train_test_split(X, Y, test_size=0.2, random_state=32)
print(xtr.shape, ytr.shape)
print(xts.shape, yts.shape)
```

```python
gs.fit(xtr, ytr)
gs.best_params_
```

```python
f1_score(gs.predict(xts), yts, average='macro')
```

```python
X.shape
```

```python
cls = RandomForestClassifier(n_jobs=4, min_samples_split=8, n_estimators=120)
cls.fit(X, Y)
pred = cls.predict(ts.drop(['age', 'disease', 'diseasecode'], axis=1).values)
```

```python
dct = dict( enumerate(d['disease'].cat.categories) )
dct
```

```python
preds = [dct[p] for p in pred]
```

```python
final = pd.DataFrame({'patientID': test['patientID'],
              'Disease': preds})
```
