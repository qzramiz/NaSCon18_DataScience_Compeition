```python
import pandas as pd
import numpy as np
pd.options.display.max_columns =100
```

```python
er = pd.read_csv('engineered_rules.csv')
print(er.shape)
print(er.isnull().sum())
er.head()
```

```python
train = pd.read_csv('Patient_Records_Train.csv')
test = pd.read_csv('Patient_Records_Test.csv')
print(train.shape)
print(test.shape)
train['from'] = 'train'
test['from'] = 'test'
all_ = pd.concat([train, test.drop('Usage', axis=1)])
all_.columns = all_.columns.str.lower()
print(all_.shape)
all_.columns
```

```python
all_.head(3)
```

```python
all_.columns = all_.columns.str.replace(' ', '_')
```

```python
all_.isnull().sum()
```

```python
all_ = all_.drop(['acute_infarction_(localization)',
           'additional_diagnoses',
           'admission_date',
           'age',
           'previous_infarction_(1)_date',
           'previous_infarction_(2)_date',
           'sex',
           'ventriculography'], axis=1)
```

```python
all_.columns
```

```python
dates = all_.loc[:,all_.columns.str.contains('date')].fillna(False).astype(bool).astype(int)
dates.columns = dates.columns.str.replace('_date', '_done')
all_ = all_.drop(all_.columns[all_.columns.str.contains('date')], axis=1)
dates.head()
```

```python
er.shape, dates.shape
```

```python
er = er.join(dates.reset_index().drop('index', axis=1))
```

```python
all_.isnull().sum()
```

```python
cats = all_.columns[all_.dtypes == 'object']
nums = all_.columns[~(all_.dtypes == 'object')]
cats, nums
```

```python
cats = all_[cats]
nums = all_[nums]
```

```python
nums
```

```python
means = nums.loc[:,nums.columns.str.contains('mean')]
means.head()
```

```python
am = means['aorta_(at_rest)_mean'].median()
papt = means['aorta_(at_rest)_mean'].median()
papl = means['aorta_(at_rest)_mean'].median()
```

```python
nums['aorta_(at_rest)_diast'].fillna(am, inplace=True)
nums['aorta_(at_rest)_syst'].fillna(am, inplace=True)
```

```python
nums['pulmonary_artery_pressure_(at_rest)_diast'].fillna(papt, inplace=True)
nums['pulmonary_artery_pressure_(at_rest)_syst'].fillna(papt, inplace=True)
```

```python
nums['pulmonary_artery_pressure_(laod)_diast'].fillna(papl, inplace=True)
nums['pulmonary_artery_pressure_(laod)_syst'].fillna(papl, inplace=True)
```

```python
nums['pulmonary_artery_pressure_(at_rest)_(mean)']
```

```python
nums = nums.drop(['aorta_(at_rest)_mean', 'pulmonary_artery_pressure_(laod)_(mean)', 'pulmonary_artery_pressure_(at_rest)_(mean)'], axis=1)
```

```python
for col in (nums.isnull().sum() > 0).index:
    nums[col].fillna(nums[col].median(), inplace=True)
nums.isnull().sum()
```

```python
nums = nums.drop('chest_x-ray', axis=1)
```

```python
er = er.join(nums.reset_index().drop('index', axis=1))
```

```python
for col in cats.columns:
    onehot = cats[col].str.get_dummies()
    onehot.columns += '_{}'.format(col)
    er = er.join(onehot.reset_index().drop('index', axis=1))
```
